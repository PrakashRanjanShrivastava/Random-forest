```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import statsmodels.formula.api as smf
import io
```

```python
from google.colab import files
data_to_load = files.upload()
```

```python
diab_data = pd.read_csv(io.BytesIO(data_to_load['diabetes.csv']))
```

```python
diab_data.head(10)
```

```python
diab_data.dtypes
```

```python
diab_data['diabetes'] = diab_data['diabetes'].map({'neg':0 , 'pos':1})
diab_data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = diab_data.iloc[:,[0,7]].values
y= diab_data.iloc[:,8]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.25,random_state = 20)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
x_train= sc_x.fit_transform(x_train)
x_test = sc_x.fit_transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rt_model = RandomForestClassifier(criterion = 'entropy',n_estimators = 2,random_state = 0)
```

```python
rt_model = rt_model.fit(x_train,y_train)
```

```python
y_pred = rt_model.predict(x_test)
y_pred
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
accuracy  = accuracy_score(y_test,y_pred)
print(accuracy)
```

```python
from sklearn.metrics import classification_report
cr =classification_report(y_test,y_pred)
print(cr)
```
